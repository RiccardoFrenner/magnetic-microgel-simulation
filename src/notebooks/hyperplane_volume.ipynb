We investigate a different method to compute the "size"/"extend" of microgels via a hyperplane based approach.

The idea is as follows:
- we specify a number of directions n
- the specific direction vectors are then drawn randomly from the unit hypersphere (optionally we can use sobol sampling to improve coverage of the space)
- for each direction we then move a hyperplan with normal vector equal to the direction vector until a certain amount of percent of points is "behind" the plane
- when done for each direction, we compute the volume of the gel as the volume enclosed by the hyperplanes
